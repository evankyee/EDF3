```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

orig= pd.read_csv('orig.csv', header=0)
mod= pd.read_csv('mod.csv', header=0)
ogtime= pd.read_csv('ogtime.csv', header=0)
modtime= pd.read_csv('modtime.csv', header=0)
auto= pd.read_csv('auto.csv', header=0)



candidates = orig['Candidates']
screened = orig['Screened']
interviewed = orig['Interviewed']
total = orig['Matched']

# Plot the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference

# Plot the lines for Matched, Receive Offers, and Total Time
plt.plot(candidates, screened, label='Screened')
plt.plot(candidates, interviewed, label='Interviewed')
plt.plot(candidates, total, label='Offers (Matched)')

plt.xlabel('Volume')
plt.ylabel('Candidates')
plt.title('Original Progression vs Applicant Volume')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig("orig.jpeg", format="jpeg")

candidates = mod['Candidates']
screened = mod['Screened']
total = mod['Matched']
interviewed = mod['Interviewed']


# Plot the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference

# Plot the lines for Matched, Receive Offers, and Total Time
plt.plot(candidates, screened, label='Screened')
plt.plot(candidates, total, label='Matched')
plt.plot(candidates, interviewed, label='Offers (Interviewed)')


plt.xlabel('Applicants')
plt.ylabel('Candidates')
plt.title('Modified Progression vs Applicant Volume')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig("mod.jpeg", format="jpeg")



screened = ogtime['Screen']
interview = ogtime['Interview']
matched = ogtime['Match']
total_time = ogtime['Total Time']

# Plot the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference

# Plot the lines for Matched, Receive Offers, and Total Time
plt.plot(screened, interview, label='Interview')
plt.plot(screened, matched, label='Matching')
plt.plot(screened, total_time, label='Total Time')

plt.xlabel('Applicants')
plt.ylabel('Time')
plt.title('Original Time vs Candidate Volume')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig("ogtime.jpeg", format="jpeg")



screened = modtime['Screen']
matched = modtime['Match']
interview = modtime['Interview']
total_time = modtime['Total Time']

# Plot the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference

# Plot the lines for Matched, Receive Offers, and Total Time
plt.plot(screened, matched, label='Matching')
plt.plot(screened, interview, label='Interview')
plt.plot(screened, total_time, label='Total Time')

plt.xlabel('Applicants')
plt.ylabel('Time')
plt.ylim(0, 17500)
plt.title('Modified Time vs Candidate Volume')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig("modtime.jpeg", format="jpeg")


screened = auto['Applicants']
matched = auto['Match']
interview = auto['Interview']
total_time = auto['Total Time']

# Plot the data
plt.figure(figsize=(10, 6))  # Adjust the figure size as per your preference

# Plot the lines for Matched, Receive Offers, and Total Time
plt.plot(screened, matched, label='Matching')
plt.plot(screened, interview, label='Interview')
plt.plot(screened, total_time, label='Total Time')

plt.xlabel('Applicants')
plt.ylabel('Time')
plt.ylim(0, 17500)
plt.title('Automated Time vs Candidate Volume')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig("autotime.jpeg", format="jpeg")


otot=ogtime['Total Time']
modtot=modtime['Total Time']
autotot=auto['Total Time']
plt.figure(figsize=(10, 6))

plt.plot(screened, otot, label='Original')
plt.plot(screened, modtot, label='Modified')
plt.plot(screened, autotot, label='Automated')

plt.xlabel('Applicants')
plt.ylabel('Time')
plt.title('Stratigies vs Candidate Volume')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig("alltimes.jpeg", format="jpeg")


```
